## Part 1.a
The following code segment and output are used to create and display the Bayesian Network as shown in the file Project-6-BayesianNetwork on D2L. To create this Bayesian Network, we use PyAgrum and the fastBN function present in the PyAgrum library to quickly define the required CPTs and connections between CPTs. At the same time, we also define all of the states (for example, True and False) that each CPT will have when we are defining the CPTs and connections between CPTs. Lastly, as the CPTs in the created Bayesian Network contain random data, we fill out all of the CPTs based on the data present in the file Project-6-BayesianNetwork on D2L and then we display the created Bayesian Network. 

Note: The Season CPT is created so that Season:0 refers to Low Epidemic Season and Season:1 refers to High Epidemic Season.

The code from the file Lab08-BayesianNetwork on D2L and from https://pyagrum.readthedocs.io/en/0.17.3/ was used to create the code displayed below.

```python
#Imports
import os, random
import matplotlib.pyplot as plt
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb
import numpy as np

%matplotlib inline

#Set a seed
r_seed = 101

#For reproductibility, this needs to be set
gum.initRandom(r_seed) 
random.seed(r_seed)

#Create a Bayesian Network
#Season and Flight connect to SARS and SARS connects to Fever and Cough
#Each CPT has False and True as values other than the Season CPT which has Low and High
bn1 = gum.fastBN('Season{L|H}->SARS{F|T}<-Flight{F|T};' + 'SARS->Fever{F|T};' + 'SARS->Cough{F|T}')

#Set the probabilities for the Season CPT: Low is 0.9 and High is 0.1
bn1.cpt('Season')[:] = [0.9, 0.1]

#Set the probabilities for the Flight CPT: False is 0.7 and True is 0.3
bn1.cpt('Flight')[:] = [0.7, 0.3]

#Set the probabilities for the SARS CPT, since the SARS CPT depends on Season and Flight (all 3 of which have 2 possible states),
#we have 2 * 2 * 2 = 8 possible variations and entries
bn1.cpt('SARS')[{'Season':0, 'Flight':0}] = [0.98, 0.02]
bn1.cpt('SARS')[{'Season':0, 'Flight':1}] = [0.97, 0.03]
bn1.cpt('SARS')[{'Season':1, 'Flight':0}] = [0.98, 0.02]
bn1.cpt('SARS')[{'Season':1, 'Flight':1}] = [0.95, 0.05]

#Set the probabilities for the Fever CPT, since the Fever CPT depends on SARS (both of which have 2 possible states),
#we have 2 * 2 = 4 possible variations and entries
bn1.cpt('Fever')[{'SARS':0}] = [0.8, 0.2]
bn1.cpt('Fever')[{'SARS':1}] = [0.1, 0.9]

#Set the probabilities for the Cough CPT, since the Cough CPT depends on SARS (both of which have 2 possible states),
#we have 2 * 2 = 4 possible variations and entries
bn1.cpt('Cough')[{'SARS':0}] = [0.7, 0.3]
bn1.cpt('Cough')[{'SARS':1}] = [0.35, 0.65]

#Display the CPT tables and the Bayesian Network that we have created
gnb.sideBySide(bn1.cpt('Season'), bn1.cpt('Flight'),bn1.cpt('SARS'), bn1.cpt('Fever'), bn1.cpt('Cough'))
gnb.showInference(bn1)
```

## Part 1.b
The following code segment and output displayed are for the scenario in which we want to find the probability of an individual having SARS given that the individual has a Fever but does not have a Cough. This uses the code displayed in Part 1.a and expands on it by creating evidence and propagating through the Bayesian Network we had created before with this new evidence in order to update the probabilities of the rest of the CPTs in the Bayesian Network. For this scenario, the evidence that we set is that Fever = True and Cough = False. We then specifically query for the probabilities of the SARS CPT and obtain the probability of an individual having SARS given that they have a Fever but no Cough, or in the other words the probability P(SARS = True | Fever = True, Cough = False).

The code from the file Lab08-BayesianNetwork on D2L and from the website https://pyagrum.readthedocs.io/en/latest/notebooks/41-Inference_graphicalInference.html was used to create the code displayed below.

```python
#Imports
import os, random
import matplotlib.pyplot as plt
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb
import numpy as np

%matplotlib inline

#Set a seed
r_seed = 101

#For reproductibility, this needs to be set
gum.initRandom(r_seed) 
random.seed(r_seed)

#Create a Bayesian Network
#Season and Flight connect to SARS and SARS connects to Fever and Cough
#Each CPT has False and True as values other than the Season CPT which has Low and High
bn1 = gum.fastBN('Season{L|H}->SARS{F|T}<-Flight{F|T};' + 'SARS->Fever{F|T};' + 'SARS->Cough{F|T}')

#Set the probabilities for the Season CPT: Low is 0.9 and High is 0.1
bn1.cpt('Season')[:] = [0.9, 0.1]

#Set the probabilities for the Flight CPT: False is 0.7 and True is 0.3
bn1.cpt('Flight')[:] = [0.7, 0.3]

#Set the probabilities for the SARS CPT, since the SARS CPT depends on Season and Flight (all 3 of which have 2 possible states),
#we have 2 * 2 * 2 = 8 possible variations and entries
bn1.cpt('SARS')[{'Season':0, 'Flight':0}] = [0.98, 0.02]
bn1.cpt('SARS')[{'Season':0, 'Flight':1}] = [0.97, 0.03]
bn1.cpt('SARS')[{'Season':1, 'Flight':0}] = [0.98, 0.02]
bn1.cpt('SARS')[{'Season':1, 'Flight':1}] = [0.95, 0.05]

#Set the probabilities for the Fever CPT, since the Fever CPT depends on SARS (both of which have 2 possible states),
#we have 2 * 2 = 4 possible variations and entries
bn1.cpt('Fever')[{'SARS':0}] = [0.8, 0.2]
bn1.cpt('Fever')[{'SARS':1}] = [0.1, 0.9]

#Set the probabilities for the Cough CPT, since the Cough CPT depends on SARS (both of which have 2 possible states),
#we have 2 * 2 = 4 possible variations and entries
bn1.cpt('Cough')[{'SARS':0}] = [0.7, 0.3]
bn1.cpt('Cough')[{'SARS':1}] = [0.35, 0.65]

#Create evidence. In this case, we have the evidence that Fever is True and that Cough is False
evidence = {'Fever': True, 'Cough': False}

#Use the Lazy Propagation algorithm present in the PyAgrum library to make inferences 
inf = gum.LazyPropagation(bn1)

#Set the evidence we created earlier
inf.setEvidence(evidence)

#Make an inference based off of the evidence that we have set
inf.makeInference()

#Having set the evidence, display our Bayesian Network given the evidence that we have set
gnb.showInference(bn1, evs = evidence)

#Specifically, grab the updated SARS probabilities based off of our evidence
res_SARS = inf.posterior('SARS')

#Print the probability of SARS being True given that the evidence is Fever = True and Cough = False
print('The probability of an individual having SARS given that the individual has a Fever and does not have a Cough is: %0.4f%%.' % (res_SARS[1] * 100))
```

## Part 2.a
The following code segment and output displayed are of the Bayesian Network from Part 1.a which has then been slightly modified. Instead of simply having two states, the Season CPT has now been declared to have 3 states which are: Low Epidemic, Medium Epidemic, and High Epidemic. Although this has obviously affected the Season CPT (as it now has 3 states), it has also affected the SARS CPT as the SARS CPT relies on the Season CPT. As a result, the SARS CPT now has 3 * 2 * 2 = 12 entries as the Season CPT now has 3 states, the Flight CPT has 2 states, and the SARS CPT has 2 states. As there are now more entries in the Season CPT and the SARS CPT than in Part 1.a, we have to create new data to fill all of the entries. 

For the Season CPT, I assumed that the addition of a Medium Epidemic state would most likely change what constitues the boundaries of what is considered a High Epidemic Season and what is considered a Low Epidemic Season. Therefore, both the High Epidemic Season and Low Epidemic Season probability would be affected in order to accomodate for the new Medium Epidemic Season. In doing so, I also assumed that the Low Epidemic Season probability would most likely be more affected than the High Epidemic Season probability because of how great the Low Epidemic Season probability is. Therefore, I decided to set the probabilities in the Season CPT as such: Low = 0.56, Medium = 0.37, and High = 0.07. 

For the SARS CPT, we also have to accomodate for the new Medium Epidemic Season state. By looking at the probabilities, it can be seen that regardless of whether the Season is Low Epidemic or High Epidemic, as long as Flight is False (the Flight is from a non-epidemic area), the probability of SARS being False is 0.98 and the probability of SARS being True is 0.02. Therefore, the new Medium Epidemic Season state would also not affect the probability of SARS being False or True greatly and it would mainly be dependent on the state of where the Flight was from. As a result, we can set the states to P(SARS = False | Season = Medium, Flight = False) = 0.98 and P(SARS = True | Season = Medium, Flight = False) = 0.02. When Flight is True (or from an epidemic area), the state of Season does affect the probability of SARS being True or False. It can be seen that during Low Epidemic Season, when Flight is True, the probability of SARS being False is 0.97 and during High Epidemic Season this probability is 0.95. Therefore, it would make sense to set the probability of SARS being False during Medium Epidemic Season, and when Flight is True, between those two numbers. As a result, we set the states to P(SARS = False | Season = Medium, Flight = True) = 0.96 and, correspondingly, P(SARS = True | Season = Medium, Flight = True) = 0.04. 

Note: In this case, as the Season CPT now has 3 states, the Season CPT is created such that Season:0 refers to Low Epidemic Season, Season:1 refers to Medium Epidemic Season, and Season:2 refers to High Epidemic Season.

The code from the file Lab08-BayesianNetwork on D2L and from https://pyagrum.readthedocs.io/en/0.17.3/ was used to create the code displayed below.
